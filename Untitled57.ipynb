```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
from scipy.integrate import simps

# Define parameters
t = 0.0
v = 1
w = 1.2

# Define the Hamiltonian matrix as a function of kx and ky
def hamiltonian(kx, ky):
    h_matrix = np.array([[2 * t * np.cos(ky), v + w * np.exp(-1j * kx)],
                         [v + w * np.exp(1j * kx), 2 * t * np.cos(ky)]])
    return h_matrix
```

```python
def berry_connection(kx, ky):
    # Calculate the Berry connection A(k)
    h_matrix = hamiltonian(kx, ky)
    eigenvalues, eigenvectors = eigh(h_matrix)
    
    # Assuming the system is in 2D, consider the lower band
    lower_band = eigenvectors[:, 1]
    
    # Calculate the Berry connection
    A_kx = 1j * np.sum(lower_band.conj() * np.gradient(lower_band, kx))
    A_ky = 1j * np.sum(lower_band.conj() * np.gradient(lower_band, ky))
    
    return A_kx, A_ky

def berry_curvature(kx_vals, ky_vals):
    # Calculate Berry curvature from the Berry connection
    A_kx_vals, A_ky_vals = np.zeros((len(kx_vals), len(ky_vals))), np.zeros((len(kx_vals), len(ky_vals)))
    
    for i, kx in enumerate(kx_vals):
        for j, ky in enumerate(ky_vals):
            A_kx, A_ky = berry_connection(kx, ky)
            A_kx_vals[i, j] = A_kx
            A_ky_vals[i, j] = A_ky
    
    # Calculate Berry curvature components
    berry_curvature_kx = np.gradient(A_ky_vals, kx_vals, axis=1) - np.gradient(A_kx_vals, ky_vals, axis=0)
    berry_curvature_ky = np.gradient(A_kx_vals, ky_vals, axis=0) - np.gradient(A_ky_vals, kx_vals, axis=1)
    
    return berry_curvature_kx, berry_curvature_ky

def polarization_from_curvature(kx_vals, ky_vals):
    # Calculate Berry curvature from the Berry connection
    berry_curvature_kx, berry_curvature_ky = berry_curvature(kx_vals, ky_vals)
    
    # Double integrals to obtain polarization
    total_polarization_x = -simps(simps(berry_curvature_kx, ky_vals), kx_vals) / (2 * np.pi)
    total_polarization_y = -simps(simps(berry_curvature_ky, kx_vals), ky_vals) / (2 * np.pi)
    
    return total_polarization_x, total_polarization_y, berry_curvature_kx, berry_curvature_ky

```

```python
def berry_phase_component(k_vals, A_vals):
    # Calculate the Berry phase component for a given k
    return np.angle(np.exp(1j * simps(A_vals, k_vals)))

def berry_phase(kx_vals, ky_vals):
    # Calculate the Berry phase for both kx and ky
    A_kx_vals, A_ky_vals = np.zeros((len(kx_vals), len(ky_vals))), np.zeros((len(kx_vals), len(ky_vals)))
    
    for i, kx in enumerate(kx_vals):
        for j, ky in enumerate(ky_vals):
            A_kx, A_ky = berry_connection(kx, ky)
            A_kx_vals[i, j] = A_kx
            A_ky_vals[i, j] = A_ky
    
    berry_phase_kx = berry_phase_component(kx_vals, A_kx_vals)
    berry_phase_ky = berry_phase_component(ky_vals, A_ky_vals)
    
    # Total Berry phase is the sum of individual components
    total_berry_phase = berry_phase_kx + berry_phase_ky
    
    return berry_phase_kx, berry_phase_ky, total_berry_phase
```

```python
import numpy as np
import sympy as sp
from scipy.linalg import eigh
from scipy.integrate import simps
import matplotlib.pyplot as plt

# Parameters
v = 0.5
w = 0.2
ky_vals = np.linspace(- 2*np.pi,  2*np.pi, 50)
kx_vals = np.linspace(- 2*np.pi,  2*np.pi, 50)

# Define symbolic variables
kx, ky = sp.symbols('kx ky', real=True)
H = sp.Matrix([
    [0, v + w * sp.exp(-1j * kx), v + w * sp.exp(-1j * ky), 0],
    [v + w * sp.exp(1j * kx), 0, 0, v + w * sp.exp(-1j * ky)],
    [v + w * sp.exp(1j * ky), 0, 0, v + w * sp.exp(-1j * kx)],
    [0, v + w * sp.exp(1j * ky), v + w * sp.exp(1j * kx), 0]
])


# Function to calculate Berry curvature components for all bands
def berry_curvature(kx_vals, ky_vals):
    berry_curvature_kx_vals = np.zeros((4, len(kx_vals), len(ky_vals)))
    berry_curvature_ky_vals = np.zeros((4, len(kx_vals), len(ky_vals)))

    for i, kx_val in enumerate(kx_vals):
        for j, ky_val in enumerate(ky_vals):
            # Substitute numerical values into the Hamiltonian matrix
            H_numerical = H.subs({kx: kx_val, ky: ky_val})

            # Diagonalize the Hamiltonian matrix
            _, eigenvectors = np.linalg.eigh(np.array(H_numerical).astype(np.complex128))

            # Loop over all bands
            for band_index in range(4):
                eigenvector = eigenvectors[:, band_index]

                # Berry connection components
                A_kx = np.imag(np.conj(eigenvector).T @ np.gradient(eigenvector, kx_val))
                A_ky = np.imag(np.conj(eigenvector).T @ np.gradient(eigenvector, ky_val))

                # Store the Berry connection components for each band at each point
                berry_curvature_kx_vals[band_index, i, j] = A_kx
                berry_curvature_ky_vals[band_index, i, j] = A_ky

    # Calculate Berry curvature components
    berry_curvature_kx = np.gradient(berry_curvature_ky_vals, axis=1, edge_order=2) - np.gradient(berry_curvature_kx_vals, axis=0, edge_order=2)
    berry_curvature_ky = np.gradient(berry_curvature_kx_vals, axis=0, edge_order=2) - np.gradient(berry_curvature_ky_vals, axis=1, edge_order=2)

    return berry_curvature_kx, berry_curvature_ky



# Calculate Berry curvature for all bands
berry_curvature_kx, berry_curvature_ky = berry_curvature(kx_vals, ky_vals)


# Function to calculate polarization from curvature for all bands
def polarization_from_curvature(kx_vals, ky_vals):
    # Calculate Berry curvature from the Berry connection for all bands
    berry_curvature_kx, berry_curvature_ky = berry_curvature(kx_vals, ky_vals)
    
    total_polarization_x_vals = np.zeros(4)
    total_polarization_y_vals = np.zeros(4)

    for band_index in range(4):
        # Double integrals to obtain polarization
        total_polarization_x = -simps(simps(berry_curvature_kx[band_index], ky_vals), kx_vals) / (2 * np.pi)
        total_polarization_y = -simps(simps(berry_curvature_ky[band_index], kx_vals), ky_vals) / (2 * np.pi)

        total_polarization_x_vals[band_index] = total_polarization_x
        total_polarization_y_vals[band_index] = total_polarization_y

    return total_polarization_x_vals, total_polarization_y_vals

# Calculate polarization for all bands
total_polarization_x_vals, total_polarization_y_vals = polarization_from_curvature(kx_vals, ky_vals)



# Plot polarization components and magnitude
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Plot polarization components
for band_index in range(4):
    axes[0, 0].plot(band_index + 1, total_polarization_x_vals[band_index], marker='o', label=f'Band {band_index + 1}')
    axes[0, 1].plot(band_index + 1, total_polarization_y_vals[band_index], marker='o', label=f'Band {band_index + 1}')

axes[0, 0].set_title('Polarization Component $P_x$')
axes[0, 0].set_xlabel('Band')
axes[0, 0].set_ylabel('$P_x$')
axes[0, 0].legend()

axes[0, 1].set_title('Polarization Component $P_y$')
axes[0, 1].set_xlabel('Band')
axes[0, 1].set_ylabel('$P_y$')
axes[0, 1].legend()

# Plot polarization magnitude
polarization_magnitude = np.sqrt(total_polarization_x_vals**2 + total_polarization_y_vals**2)
axes[1, 0].plot(range(1, 5), polarization_magnitude, marker='o')
axes[1, 0].set_title('Polarization Magnitude')
axes[1, 0].set_xlabel('Band')
axes[1, 0].set_ylabel('Magnitude')

# Plot polarization magnitude
polarization_magnitude = np.abs((total_polarization_x_vals**2 + total_polarization_y_vals**2))
axes[1, 1].plot(range(1, 5), polarization_magnitude, marker='o')
axes[1, 1].set_title('Polarization Magnitude')
axes[1, 1].set_xlabel('Band')
axes[1, 1].set_ylabel('Magnitude')


# Show the plots
plt.tight_layout()
plt.show()
```
